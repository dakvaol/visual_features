```python
import os
import pandas as pd 
import youtube_dl
import subprocess
```

```python
asset_df = pd.read_csv("../datasets/movie_assets/movie_assets_exposed.csv")
```

```python
asset_df.head()
```

```python
# Makes a list of unique assets from a series of assets
def get_assets(series):
    unique_list = []

    for x in series:
        if x not in unique_list:
            unique_list.append(x)
    return unique_list

# Get a specific title based on the movie title 
def get_asset(title):
    return asset_df['assetId'].loc[asset_df['title'] == title] 
```

```python
print(get_asset("Forever My Girl"))
```

```python
assets = get_assets(asset_df['assetId'])
print(assets)
```
